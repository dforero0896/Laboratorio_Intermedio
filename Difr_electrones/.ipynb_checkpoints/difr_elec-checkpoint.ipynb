```python
from scipy.optimize import curve_fit
plt.plot(1./np.sqrt(Volt), D1, 'or')
plt.plot(1./np.sqrt(Volt), D2, 'og')
def func(x, a):
    return a * x
x = np.linspace(0, 0.8, 100)
popt, pcov = curve_fit(func, 1./np.sqrt(Volt), D1)
plt.plot(x, func(x, *popt), 'r-', label='fit')
popt2, pcov2 = curve_fit(func, 1./np.sqrt(Volt), D2)
plt.plot(x, func(x, *popt2), 'g-', label='fit')
```
