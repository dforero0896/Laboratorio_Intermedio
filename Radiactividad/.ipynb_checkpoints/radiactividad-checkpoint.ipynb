```python
#Actividad 3
t_sat=3 #s
#con Ra
d4cm=np.array([7850., 8075, 7910, 7962., 7973])
d5cm=np.array([5687., 5761., 5936., 5804., 5736.])
d6cm=np.array([4387., 4472., 4479, 4439., 4566])
d7cm=np.array([3612., 3443., 3386., 3547, 3419.])
d85cm=np.array([2762., 2793., 2789., 2713, 2746])
d10cm=np.array([1847., 1742., 1751., 1656., 1862.])
d13cm=np.array([963., 1079., 1048, 1159., 1041.])
arr_distances=[d4cm, d5cm, d6cm, d7cm,d85cm, d10cm, d13cm]
mean_distances=[]
st_distances=[]
for i in range(len(arr_distances)):
    mean_distances.append(np.mean(arr_distances[i]))
    #st_distances.append(np.std(arr_distances[i]))
    st_distances.append(np.sqrt(np.mean(arr_distances[i])))
mean_distances=np.array(mean_distances)
distances=np.array([4., 5, 6, 7,8, 10, 13])
st_distances=np.array(st_distances)
```

```python
plt.errorbar(distances, mean_distances/1000, yerr=st_distances/1000, marker='o', linewidth=0)
from scipy.optimize import curve_fit
r=np.linspace(0, 15, 1000)
def func(x, a, b, c, d):
    return a*np.exp(-b*x)+d

popt, pcov = curve_fit(func, distances, mean_distances/1000)
plt.plot(r, func(r,*popt))
```

```python
print 0.5*max(mean_distances/1000)
print func(6.448, *popt)
```

```python
plt.errorbar(distances**(-2), mean_distances, yerr=st_distances, marker='o', linewidth=0)
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(distances**(-2), mean_distances)

rs = np.linspace(0, max(distances**(-2)), 1000000)
plt.plot(rs, slope*rs+intercept)
```
