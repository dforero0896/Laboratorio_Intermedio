```python
def get_rad_perc(nc, pc, Pbc):
    tot = np.mean(nc)
    beta = np.mean(pc)-np.mean(Pbc)
    alpha = np.mean(nc)-np.mean(pc)
    gamma = np.mean(Pbc)
    return 100.*alpha/tot, 100.*beta/tot, 100.*gamma/tot
    
```

```python
def plot_material(material_dict, title, i, k):
    if(material_dict=={1:2}):
        return 0
    else:
        x, y = zip(*sorted(material_dict.items()))
        wid = np.linspace(min(x), max(x), 1000)
        #plt.plot(x,np.log10(y), 'o')
        ax[i, k].plot(x,y, 'o', label='data')
        ax[i, k].set_xlabel('Grosor (in)')
        ax[i, k].set_ylabel('$N$', fontsize=15)
        ax[i, k].set_yscale('log')
        #plt.ylim(100, 1500)
        def func(x, a, b):
            return a*np.exp(-b*x)
        #popt, pcov = curve_fit(func, x, np.log10(y))
        popt, pcov = curve_fit(func, x, y)
        #plt.plot(wid, func(wid,*popt), label='fit')
        ax[i, k].plot(wid, func(wid,*popt), label='fit')
        f = lambda d : func(d,*popt) - 0.5*max(y)
        x_sol = fsolve(f, 0.04)
        x_sol
        ax[i, k].set_title(title+', mu =%.2f, x50 = %.2f'%(popt[1], x_sol))
        #ax[i, k].xaxis.set_major_formatter(ticker.FormatStrFormatter('%0.2f'))
        print title+'&$%.2f \pm %.2f$&%.2f'%(popt[1],np.sqrt(np.diag(pcov))[0], x_sol)
```
